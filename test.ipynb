```python
import pandas as pd
```

```python
data = pd.read_csv('database.csv')
data
```

```python
data['month'] = data['departure'].apply(lambda ts: ts.split('-')[0] + '-' + ts.split('-')[1])
data['month'].unique()
```

```python
for i in range(data['month'].unique().__len__()):
    data[data['month'] == data['month'].unique()[i]].to_csv('./split_data/split_data_' + data['month'].unique()[i] + '.csv', index=False)
```
